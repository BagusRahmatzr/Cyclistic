# Fase Pemrosesan 

```python
#meinstal library yang mungkin diperlukan 
library(tidyverse)
library(lubridate)
library(ggplot2)
library(anytime)
library(dplyr)
library(skimr)
library(janitor)

#Mengimport Data
tripdata_202005 <- read.csv('D:\\Data Sepeda Cyclistic\\202005-divvy-tripdata.csv')
tripdata_202006 <- read.csv('D:\\Data Sepeda Cyclistic\\202006-divvy-tripdata.csv')
tripdata_202007 <- read.csv('D:\\Data Sepeda Cyclistic\\202007-divvy-tripdata.csv')
tripdata_202008 <- read.csv('D:\\Data Sepeda Cyclistic\\202008-divvy-tripdata.csv')
tripdata_202009 <- read.csv('D:\\Data Sepeda Cyclistic\\202009-divvy-tripdata.csv')
tripdata_202010 <- read.csv('D:\\Data Sepeda Cyclistic\\202010-divvy-tripdata.csv')
tripdata_202011 <- read.csv('D:\\Data Sepeda Cyclistic\\202011-divvy-tripdata.csv')
tripdata_202012 <- read.csv('D:\\Data Sepeda Cyclistic\\202012-divvy-tripdata.csv')
tripdata_202101 <- read.csv('D:\\Data Sepeda Cyclistic\\202101-divvy-tripdata.csv')
tripdata_202102 <- read.csv('D:\\Data Sepeda Cyclistic\\202102-divvy-tripdata.csv')
tripdata_202103 <- read.csv('D:\\Data Sepeda Cyclistic\\202103-divvy-tripdata.csv')
tripdata_202104 <- read.csv('D:\\Data Sepeda Cyclistic\\202104-divvy-tripdata.csv')

#melihat nama variabel/Kolom yang ada dalam data
colnames(tripdata_202005)
colnames(tripdata_202006)
colnames(tripdata_202007)
colnames(tripdata_202008)
colnames(tripdata_202009)
colnames(tripdata_202010)
colnames(tripdata_202011)
colnames(tripdata_202012)
colnames(tripdata_202101)
colnames(tripdata_202102)
colnames(tripdata_202103)
colnames(tripdata_202104)

#mengetahui struktur dari data
str(tripdata_202005)
str(tripdata_202006)
str(tripdata_202007)
str(tripdata_202008)
str(tripdata_202009)
str(tripdata_202010)
str(tripdata_202011)
str(tripdata_202012)
str(tripdata_202101)
str(tripdata_202102)
str(tripdata_202103)
str(tripdata_202104)

#Tahap Proses
#Menggabungkan Seluruh Data
tripdata_combine_v1 <- rbind(tripdata_202005, tripdata_202006, tripdata_202007, 
                         tripdata_202008, tripdata_202009, tripdata_202010, 
                         tripdata_202011, tripdata_202012, tripdata_202101, 
                         tripdata_202102, tripdata_202103, tripdata_202104)

# memeriksa jumlah total dari baris dan kolom hasil dari menggabungkan seluruh data
dim(tripdata_combine_v1) 
#memeriksan tipe data dan nama kolom
str(tripdata_combine_v1) 

# mengubah tipe data started_ad dari chr menjadi POSIXCT(datetime)
tripdata_combine_v1$started_at <- ymd_hms(tripdata_combine_v1$started_at)

# mengubah tipe data ended_ad dari chr menjadi POSIXCT(datetime)
tripdata_combine_v1$ended_at <- ymd_hms(tripdata_combine_v1$ended_at)

# membuat kolom ridelength dalam menit dari total durasi
tripdata_combine_v2 <- tripdata_combine_v1 %>% 
  mutate(ride_length = signif(difftime(ended_at, started_at, units = 'mins'),2))

# membuat kolom day_of_week untuk menunjukan kapan pengendara mulai pada satu minggu
tripdata_combine_v2 <- tripdata_combine_v2 %>% 
  mutate(day_of_week = weekdays(started_at))

#membuat kolom bulan untuk melihat bulan berapa pengendara mulai menggunakan cyclistic
tripdata_combine_v2 <- tripdata_combine_v2 %>%
  mutate(month = format(as.Date(started_at), "%B"))

#Cleaning / Pembersihan data

# data baru untuk menghapus panjang perjalanan pengguna yang memiliki nilai negatif. menghapus perjalanan dibawah 1 menit dan diatas 24 jam untuk outlier 
tripdata_combine_v3 <- tripdata_combine_v2[(tripdata_combine_v2$ride_length > 1 & tripdata_combine_v2$ride_length < 1440), ]

# melakukan pengecekan apakah ada id pengguna yang ganda/double
any(duplicated(tripdata_combine_v3$ride_id))

# melakukan pengecekan apakah ada id pengguna yang hilang
sum(is.na(tripdata_combine_v3$ride_id))

# # melakukan pengecekan apakah ada ride_length yang hilang
sum(is.na(tripdata_combine_v3$ride_length))

# melakukan pengecekan member_casual hanya untuk dua tipe entri
table(tripdata_combine_v3$member_casual)

# melakukan pengecekan member_casual hanya untuk tiga tipe entri
table(tripdata_combine_v3$rideable_type)

# Ordering Days of the week
tripdata_combine_v3$day_of_week <- ordered(tripdata_combine_v3$day_of_week, level = c("Monday", 
                                                                                      "Tuesday", 
                                                                                      "Wednesday", 
                                                                                      "Thursday", 
                                                                                      "Friday", 
                                                                                      "Saturday", 
                                                                                      "Sunday"))

# Ordering months
tripdata_combine_v3$month <- ordered(tripdata_combine_v3$month, level = c("January", 
                                                                          "February", 
                                                                          "March", 
                                                                          "April", 
                                                                          "May", 
                                                                          "June", 
                                                                          "July", 
                                                                          "August", 
                                                                          "September", 
                                                                          "October", 
                                                                          "November", 
                                                                          "December"))

# melakukan pengecekan order dalam minggu
levels(tripdata_combine_v3$day_of_week)

# melakukan pengecekan order dalam bulan
levels(tripdata_combine_v3$month)

# melihat rangkuman datafreme
glimpse(tripdata_combine_v3)
```

# Tahap Analisis

```python
#Analisis Proses

# menghitung mean, median, max, dan min untuk semua kolom ride_length
tripdata_combine_v3 %>%
  summarise(min_ride_length = min(ride_length), max_ride_length = max(ride_length), median_ride_length = median(ride_length), mean_ride_length = mean(ride_length))

# Membandingkan ukuran tendensi sentral untuk pengendara biasa dan anggota
tripdata_combine_v3 %>% 
  group_by(member_casual) %>% 
  summarise(min_length_minutes = min(ride_length), max_length_minutes = max(ride_length), 
            median_length_minutes = median(ride_length), mean_length_minutes = mean(ride_length))

table(tripdata_combine_v3$member_casual)

# menghitung rata panjang perjalanan untuk pengendara biasa dan member tahunan perhari
aggregate(tripdata_combine_v3$ride_length ~ tripdata_combine_v3$member_casual + tripdata_combine_v3$day_of_week, FUN=mean)

#menghitung total pengendara biasa dan pengendara tahunan perhari
tripdata_combine_v3 %>%
  group_by(member_casual, day_of_week) %>%
  summarise(number_of_ride = n(), .groups = 'drop') %>%
  arrange(day_of_week)

# menghitung rata rata panjang perjalanan pengendara biasa dan member tahunan perbulan
aggregate(tripdata_combine_v3$ride_length ~ tripdata_combine_v3$member_casual + tripdata_combine_v3$month, FUN=mean)

# menghitung total pengendara biasa dan member tahuan pernbulan
tripdata_combine_v3 %>%
  group_by(member_casual, month) %>%
  summarise(number_of_ride = n(), .groups = 'drop') %>%
  arrange(month)

tripdata_combine_v3 %>%
  group_by(rideable_type, member_casual) %>%
  summarise(number_of_rides = n(), .groups = 'drop')

# menExport data hasil proses dan cleaning
write.csv(tripdata_combine_v3, "D:\\Data Sepeda Cyclistic\\tripdata_fix_tableau.csv", row.names = FALSE)
```
